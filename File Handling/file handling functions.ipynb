```python
#read
with open ("example.txt", "r") as file:
    content= file.read()
    print(content)
    
```

```python
with open('example.txt', 'r') as file:
    for line in file :
        print(line.strip())
```

```python
#writing 
with open('example.txt', 'w') as file:
    file.write('hello world!')
```

```python
#append
with open('example.txt', 'a') as file:
    file.write(" \n this is a append operation!")
```

```python
#write list of line
lines=['first line \n','second line\n',' third line']
with open ('example.txt','a') as file:
    file.writelines(lines)
```

```python
#read from one file and write in other
with open('example.txt','r') as file :
    content =file.read()
with open("exm.txt",'w')as f1:
    f1.writelines(content)
```
