```python
arr=np.array([[1,2,3],[1,2,3]])
print('Array : \n', arr)
print('shape:\n' ,arr.shape)
print('number ofdimension :\n' , arr.ndim)
print("size of number of elements: \n" , arr.size)
print('data type:\n' , arr.dtype)
print('item size in bytes: \n',  arr.itemsize )
```

```python
arr1=np.array([1,2,3,])
arr2=np.array([2,3,4])
print("addition:" , arr1+arr2)
print('sub:', arr1-arr2)
print('div:', arr1 /arr2)
print('multi:', arr1*arr2)
```

```python
print(np.sqrt(arr1))
print(np.log(arr1))
```
