```python
#class
class car():
    pass
audi=car()
bmw=car()
print(audi)
audi.window=4
bmw.doors=4
print(bmw.doors)
```

```python
#instance variable and method
class Dog():
    #costructor
    def __init__ (self,name,age):
        self.name=name
        self.age=age
dog1=Dog('lucky',4)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
class Dog():
    #costructor
    def __init__ (self,name,age):
        self.name=name
        self.age=age
    def bark(self):
        print(self.name,' says woof')
dog1=Dog('lucky',4)

dog1.bark()
print(dog1)
```

```python
class student():
    def __init__( self,name, roll_no):
        self.name=name
        self.roll_no=roll_no
student1=student('sakshi',1)
student2=student('mohit',2)
print(student2.name)
```

```python
class Employee():
    def __init__(self,emp_id , name, salary):
        self.emp_id = emp_id
        self.name=name
        self.salary=salary
emp1= Employee(121,'sunita',2000 )
print(emp1.salary)
```
