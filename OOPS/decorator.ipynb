```python
#closures = function insde a functions
def mainwelcome():
    msg="welcome"
    def subwelcome():
        print("welcome to python")
        print(msg)
        print("please learn properly")

    return subwelcome()
```

```python
mainwelcome()
```

```python
#decorators
def mainwelcome(func):
    msg="welcome"
    def subwelcome():
        print("welcome to python")
        print(func())
        print("please learn properly")

    return subwelcome()
```

```python
def courseintroduction():
    return " welcome to advanced python topic "
```

```python
mainwelcome(courseintroduction)
```

```python
@mainwelcome
def  courseintroduction():
    return " welcome to advanced python topic "
```

```python
def my_decorator(func):
    def wapper():
        return " this message is before the function is call"
        print(func())
        return " this  after the function is called"
    return wapper()
```

```python
@my_decorator
def myfunction():
    return "hello this is my function"
```
